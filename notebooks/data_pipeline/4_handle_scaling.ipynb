# 4. Scaling Continuous Features

This notebook standardizes the continuous numeric features in the encoded dataset using `StandardScaler` from scikit-learn.

## 1. Imports & Setup

```python
import os
import pandas as pd
from sklearn.preprocessing import StandardScaler

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.4f}'.format)

print('Imports successful.')
```

## 2. Load Encoded Data

```python
INPUT_PATH = '../../dataset/processed/credit_card_fraud_encoded.csv'

df = pd.read_csv(INPUT_PATH)
print(f'Loaded: {df.shape[0]:,} rows x {df.shape[1]} columns')
df.head()
```

## 3. Select Columns to Scale

```python
continuous_cols = ['amount', 'amount_log', 'velocity_last_24h', 'velocity_last_24h_log', 'city_population', 'city_population_log']
scale_cols = [col for col in continuous_cols if col in df.columns]

print('Continuous columns selected for scaling:', scale_cols)
```

## 4. Apply Standard Scaling

```python
df_scaled = df.copy()

scaler = StandardScaler()
df_scaled[scale_cols] = scaler.fit_transform(df_scaled[scale_cols])

print(f'Scaled {len(scale_cols)} continuous column(s) with StandardScaler.')
df_scaled.head()
```

## 5. Verify the Transformation

```python
print('Dtype summary after scaling:')
print(df_scaled.dtypes.value_counts())

print('\nScaled columns summary statistics:')
print(df_scaled[scale_cols].describe().round(4))
```

## 6. Save Outputs

```python
OUTPUT_PATH = '../../dataset/processed/credit_card_fraud_scaled.csv'

output_dir = os.path.dirname(OUTPUT_PATH)
os.makedirs(output_dir, exist_ok=True)
df_scaled.to_csv(OUTPUT_PATH, index=False)

print(f'Saved scaled dataset to: {OUTPUT_PATH}')
print(f'Rows: {df_scaled.shape[0]:,} | Columns: {df_scaled.shape[1]:,}')
```
