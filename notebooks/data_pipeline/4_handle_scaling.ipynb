# Scaling Continuous Features

This notebook drops the identifier column and scales only the continuous features in the encoded dataset before SMOTENC.

## 1. Imports & Setup

```python
import os
import pandas as pd
from sklearn.preprocessing import StandardScaler

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.4f}'.format)

print('Imports successful.')
```

## 2. Load Encoded Data

```python
INPUT_PATH = 'C:/Users/2021ICTS28/Desktop/end-to-end-credit-card-fraud-detection-system/dataset/processed/credit_card_fraud_encoded.csv'

df = pd.read_csv(INPUT_PATH)
print(f'Loaded: {df.shape[0]:,} rows x {df.shape[1]} columns')
df.head()
```

## 3. Select Columns to Scale

```python
TARGET_COL = 'is_fraud'
ID_COL = 'transaction_id' if 'transaction_id' in df.columns else None

scale_cols = [col for col in ['amount', 'transaction_hour', 'velocity_last_24h', 'cardholder_age', 'device_trust_score_binned'] if col in df.columns]
if not scale_cols:
    raise ValueError('No numeric columns found to scale.')

print('Columns selected for scaling:', scale_cols)
```

## 4. Apply Standard Scaling

```python
df_scaled = df.copy()

if 'transaction_id' in df_scaled.columns:
    df_scaled = df_scaled.drop(columns=['transaction_id'])
    print('Dropped transaction_id before scaling.')

continuous_cols = ['amount', 'transaction_hour', 'velocity_last_24h', 'cardholder_age']
scale_cols = [col for col in continuous_cols if col in df_scaled.columns]
if not scale_cols:
    raise ValueError('No continuous columns found to scale.')

print('Columns selected for scaling:', scale_cols)

scaler = StandardScaler()
df_scaled[scale_cols] = scaler.fit_transform(df_scaled[scale_cols])

print(f'Scaled {len(scale_cols)} continuous column(s) with StandardScaler.')
df_scaled.head()
```

## 5. Verify the Transformation

```python
print('Dtype summary after scaling:')
print(df_scaled.dtypes.value_counts())

print()
print('Scaled columns summary:')
print(df_scaled[scale_cols].describe().round(4))
```

## 6. Save Outputs

```python
OUTPUT_PATH = 'C:/Users/2021ICTS28/Desktop/end-to-end-credit-card-fraud-detection-system/dataset/processed/credit_card_fraud_scaled.csv'

output_dir = os.path.dirname(OUTPUT_PATH)
os.makedirs(output_dir, exist_ok=True)
df_scaled.to_csv(OUTPUT_PATH, index=False)

print(f'Saved scaled dataset to: {OUTPUT_PATH}')
print(f'Rows: {df_scaled.shape[0]:,} | Columns: {df_scaled.shape[1]:,}')
```
