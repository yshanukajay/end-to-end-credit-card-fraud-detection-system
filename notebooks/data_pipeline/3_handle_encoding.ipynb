# 3. Handle Encoding

This notebook encodes nominal categorical columns (`merchant_category`, `gender`) using one-hot encoding, and encodes ordinal binned columns (`customer_age_binned`, `transaction_hour_binned`, `distance_to_merchant_binned`) using custom ordered integer mappings.

## 1. Imports & Setup

```python
import os
import pandas as pd
import numpy as np

pd.set_option('display.max_columns', None)
pd.set_option('display.float_format', '{:.4f}'.format)

print('Imports successful.')
```

## 2. Load Binned Data

```python
INPUT_PATH = '../../dataset/processed/credit_card_fraud_binned.csv'

df = pd.read_csv(INPUT_PATH)
print(f'Loaded: {df.shape[0]:,} rows x {df.shape[1]} columns')
df.head()
```

## 3. Select Columns to Encode

```python
one_hot_cols = ['merchant_category', 'gender']

label_mappings = {
    'customer_age_binned': {'Youth': 0, 'Young-Adult': 1, 'Middle-Aged': 2, 'Senior': 3},
    'transaction_hour_binned': {'Morning': 0, 'Afternoon': 1, 'Evening': 2, 'Night': 3},
    'distance_to_merchant_binned': {'Close': 0, 'Moderate': 1, 'Far': 2}
}

print('Columns selected for one-hot encoding:', one_hot_cols)
print('Label mappings for ordinal binned features:', list(label_mappings.keys()))
```

## 4. Apply Encoding

```python
df_encoded = pd.get_dummies(df, columns=one_hot_cols, drop_first=False, dtype=int)

for col, mapping in label_mappings.items():
    if col in df_encoded.columns:
        df_encoded[col] = df_encoded[col].map(mapping).astype(int)

print(f'Encoded dataset shape: {df_encoded.shape}')
df_encoded.head()
```

```python
df_encoded.columns
```

## 5. Verify the Transformation

```python
print('Dtype summary after encoding:')
print(df_encoded.dtypes.value_counts())

print('\nColumns list after encoding:')
print(list(df_encoded.columns))
```

## 6. Save Outputs

```python
OUTPUT_PATH = '../../dataset/processed/credit_card_fraud_encoded.csv'

output_dir = os.path.dirname(OUTPUT_PATH)
os.makedirs(output_dir, exist_ok=True)
df_encoded.to_csv(OUTPUT_PATH, index=False)

print(f'Saved encoded dataset to: {OUTPUT_PATH}')
print(f'Rows: {df_encoded.shape[0]:,} | Columns: {df_encoded.shape[1]:,}')
```
